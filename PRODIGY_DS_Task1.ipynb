```python
#importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading the csv file
df = pd.read_csv("population_by_age_group.csv")
```

```python
#first 5 rows of data
df.head()
```

```python
#last 5 rows of data
df.tail()
```

```python
#checking for missing values
df.isnull().sum()
```

```python
#Sum population for each age group
age_groups = ['0-4 years', '5-14 years', '15-24 years', '25-64 years', '65+']
age_group_totals = df[age_groups].sum()
```

```python
#plotting the bar chart 
age_group_totals.plot(kind='bar',grid=True)
plt.title("Population Distribution by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Population")
plt.show()
```

```python
#plotting the histogram
plt.hist(df['Total'], bins=15)
plt.title("Distribution of Total Population Across Countries")
plt.xlabel("Total Population")
plt.ylabel("Number of Countries")
plt.show()
```
